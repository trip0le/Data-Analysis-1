```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
before = data_frame['bp_before']
after = data_frame['bp_after']
age = data_frame['agegrp']
```

```python
plt.title('Age wise distribution of before blood pressure');
plt.bar(age, before);
plt.xlabel('Age Group');
plt.ylabel('Before Blood Pressure');
```

```python
plt.title('Age wise distribution of after blood pressure');
plt.bar(age, after);
plt.xlabel('Age Group');
plt.ylabel('After Blood Pressure');
```

```python
plt.bar(age, before);
plt.bar(age, after);
plt.legend(['Before', 'After']);
```

```python
sns.barplot(x='agegrp', y='bp_before', hue='sex', data = data_frame);
```

Analysis obtained :
1. **Before** and **after** treatment poses negligible effect on reducing blood pressure.
2. With age, blood pressure **increases**.
3. Males tend to have **higher** blood pressure than females.
